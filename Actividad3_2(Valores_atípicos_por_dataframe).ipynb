## **DATA 1 (2020)**

```python
data1.head()
```

```python
#Verificamos información del DataFrame
data1.info()
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data1.fillna(method="bfill") 
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia del dataframe
data2020=data1.copy()
data2020=data2.fillna(round(data1.mean(),1))
data2020
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2020.isnull().sum()
valores_nulos
```

```python
data2020nuevo = data2.select_dtypes(include = ["number"])
data2020nuevo
```

```python
dataobj2020 = data2.select_dtypes(include = ["object"])
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data2020nuevo.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data2020nuevo
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data2020nuevo[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df2020= data2020nuevo[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df2020
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=df2020.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean20=df2020.copy()
data_clean20=data_clean20.fillna(round(df2020.mean(),1))
data_clean20
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean20.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data2020nuevo

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data2020nuevo[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data2020nuevo_iqr= data2020nuevo[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2020nuevo_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data2020nuevo.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
df2020final=data2020nuevo.copy()
df2020final=df2020final.fillna(round(data2020nuevo.mean(),1))
df2020final
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=df2020final.isnull().sum()
valores_nulos
```

```python
Datos2020_limpios = pd.concat([df2020final, dataobj2020], axis=1)
Datos2020_limpios
```

```python
#Convertir DataFrame a CSV
Datos2020_limpios.to_csv("Datosgastos2020cuar_limpios.csv")
data_clean20.to_csv("Datosgastos2020des_limpios.csv")
```

```python
#descargar archivo filtrado en csv
from google.colab import files 
files.download("Datosgastos2020des_limpios.csv")
files.download("Datosgastos2020cuar_limpios.csv")
```
